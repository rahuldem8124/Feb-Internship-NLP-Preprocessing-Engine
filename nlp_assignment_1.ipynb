```python
import re
from collections import Counter

def preprocess_text(text):
    # Handle empty strings immediately
    if not text:
        return [], ""

    # Step 1: Lowercase everything
    text = text.lower()

    # Step 2: Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Step 3: Remove numbers, punctuation, and emojis (keep only letters and spaces)
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Step 4: Fix repeated characters (e.g., "looooved" -> "loved", "baaad" -> "bad")
    # This finds any character repeated 3 or more times and shrinks it down to 1
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Step 5: Tokenize by splitting on spaces
    raw_tokens = text.split()

    # Step 6: Filter out short words, but keep important exceptions
    exceptions = ['ok', 'no', 'not']
    cleaned_tokens = []

    for word in raw_tokens:
        if len(word) > 2 or word in exceptions:
            cleaned_tokens.append(word)

    # Rebuild the clean sentence
    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence


# ==========================================
# RUNNING THE ASSIGNMENT TASKS
# ==========================================

sample_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this",
    "",
    "🔥🔥🔥",
    "123 456"
]

# We will collect all tokens here for tasks 4 and 5
all_system_tokens = []

print("=== TASK 3: STRESS TESTING ===")
for sentence in sample_sentences:
    tokens, clean_text = preprocess_text(sentence)
    all_system_tokens.extend(tokens)

    print(f"Original : {sentence}")
    print(f"Tokens   : {tokens}")
    print(f"Cleaned  : {clean_text}\n")


print("=== TASK 4: TOKEN ANALYTICS ===")
total_tokens = len(all_system_tokens)
unique_tokens = len(set(all_system_tokens))

# Calculate average length safely to avoid dividing by zero
if total_tokens > 0:
    total_chars = sum(len(word) for word in all_system_tokens)
    avg_length = total_chars / total_tokens
else:
    avg_length = 0

print(f"Total number of tokens : {total_tokens}")
print(f"Number of unique tokens: {unique_tokens}")
print(f"Average token length   : {round(avg_length, 2)} characters\n")


print("=== TASK 5: FREQUENCY ANALYSIS ===")
word_frequencies = Counter(all_system_tokens)

print("Top 10 most frequent words:")
for word, count in word_frequencies.most_common(10):
    print(f" - {word}: {count}")

print("\nTop 5 least frequent words:")
# Get the least common words by slicing the end of the list and reversing it
least_frequent = word_frequencies.most_common()[-5:]
least_frequent.reverse()

for word, count in least_frequent:
    print(f" - {word}: {count}")
```
